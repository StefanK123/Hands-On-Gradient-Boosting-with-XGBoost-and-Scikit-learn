```python
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
from matplotlib.pylab import rcParams

# Sci-kit learn libraries
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.metrics import accuracy_score
from sklearn import tree
```

```python
# Import the data
df_census = pd.read_csv('census_cleaned.csv')
df_census.head()
```

```python
# Create predictor columns
X = df_census.iloc[:,:-1]
# Create target columns
y = df_census.iloc[:,-1]

# Split the data into test and training sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2)
```

```python
# Initialize the ML model (clf = classifier)
clf = DecisionTreeClassifier(max_depth=3, random_state=2)

# Fit the model on the training data set
clf.fit(X_train, y_train)

# Make predictions for the test set
y_pred = clf.predict(X_test)

# Compare predictions with the true value
print('Accuracy score:', np.round(accuracy_score(y_pred, y_test), 2))
```

```python
# Visualize the tree
# print(tree.export_text(clf))

##set the parameters of the visualisation
rcParams['figure.figsize'] = 80,50

tree.plot_tree(clf, feature_names=df_census.columns)
plt.savefig('sklearn_tree_level3.png')
#plt.show()
```

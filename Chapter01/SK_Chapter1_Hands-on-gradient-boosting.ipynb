```python
# Importing libraries
import pandas as pd
import numpy as np
import datetime as dt
```

# Data cleaning

```python
# Load data
df_bikes = pd.read_csv('bike_rentals.csv')
# Show the imported data
df_bikes.head()
```

### Explanation of dataset
+ df_bikes['cnt']: Total number of bikes rented that day (sum of registered & casual bikes)

```python
df_bikes.describe()
```

```python
df_bikes.info()
```

```python
# How many null values are in the data?
print(df_bikes.isna().sum())
print('Total number of Null values: '+str(df_bikes.isna().sum().sum()))
```

```python
# Display Null values
df_bikes[df_bikes.isna().any(axis=1)]
```

```python
# Replace Null values with median value in column Windspeed
df_bikes['windspeed'].fillna(df_bikes['windspeed'].median(), inplace=True)
```

```python
# Grouping the rental bikes by season
df_bikes.groupby(['season'])[['instant','yr','mnth','weekday','workingday','weathersit','temp', 'atemp', 'hum', \
                              'windspeed', 'casual', 'registered', 'cnt']].median()
```

```python
# Insert median humidity of each season for Null values
df_bikes['hum'] = df_bikes['hum'].fillna(df_bikes.groupby('season')['hum'].transform('median'))
```

```python
# Insert mean temperature of the day before and after
df_bikes[df_bikes['temp'].isna()] # --> Row 701 has a Null value for temperature
# Calculate mean temperatures
mean_temp = (df_bikes.iloc[700]['temp'] + df_bikes.iloc[702]['temp'])/2
mean_atemp = (df_bikes.iloc[700]['atemp'] + df_bikes.iloc[702]['atemp'])/2
# Replace the Null values
df_bikes['temp'].fillna(mean_temp, inplace=True)
df_bikes['atemp'].fillna(mean_atemp, inplace=True)
```

```python
# Fill Null values in date columns
# 1. Convert dteday column to datetime
df_bikes['dteday'] = pd.to_datetime(df_bikes['dteday'])
df_bikes['mnth'] = df_bikes['dteday'].dt.month
df_bikes.loc[730, 'yr'] = 1.0
```

```python
# Drop non-numerical values from columns
df_bikes = df_bikes.drop('dteday', axis=1)
```

```python
df_bikes.tail()
```

# Prediction with linear regression

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Drop the casual and registered bike column, because it cannot be used for prediction
df_bikes = df_bikes.drop(['casual', 'registered'], axis=1)
```

```python
# Slice the data into predictor and target column
X = df_bikes.iloc[:, :-1]
y = df_bikes.iloc[:, -1]
# Split the data into training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2)
```

```python
# 1. Initialize the model
lin_reg = LinearRegression()

# 2. Fit the model on the training set
lin_reg.fit(X_train, y_train)

# 3. Make predictions for the test set
y_pred = lin_reg.predict(X_test)

# 4. Compute mean squared error of prediction as mse
mse = mean_squared_error(y_test, y_pred)

# 5. Compute root mean squared error as rmse
rmse = np.sqrt(mse)

# Display root mean squared error
print("RSME: %0.2f" % (rmse))
```

# Cross validation with linear regression

```python
from sklearn.model_selection import cross_val_score
```

```python
# 1. Initialize machine learning model
model = LinearRegression()

# 2. Calculate cross validation score
scores = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=10)

# 3. RMSE
rmse = np.sqrt(-scores)

# 4. Display the results
print('Reg rmse:', np.round(rmse, 2))
print('RMSE mean: %0.2f' % (rmse.mean()))
```

# XGBoost

```python
from xgboost import XGBRegressor
```

```python
# 1. Initialize the model
xg_reg = XGBRegressor()

# 2. Fit the model to the training set
xg_reg.fit(X_train, y_train)

# 3. Make predictions for the test set
y_pred = xg_reg.predict(X_test)

# 4. Calculate mean squared error (mse) and root mse
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

# Print result
print("RSME: %0.2f" % (rmse))
```

# Cross validation with XGBoost

```python
# 1. Initialize the model
model = XGBRegressor()

# 2. Implement cross validation score
scores = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=10)

# 3. RMSE
rmse = np.sqrt(-scores)

# 4. Display the results
print('xg reg rmse:', np.round(rmse, 2))
print('RMSE mean: %0.2f' % (rmse.mean()))
```
